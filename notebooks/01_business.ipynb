# Proyecto: Predicción de Malas Compras en Subastas Automotrices
## Rol: Business Analyst | Sprint 1

**Dataset:** `06-kickAutomotriz.csv`  
**Fuente:** Subastas mayoristas de autos usados (EE.UU.)  
**Registros:** 72,983 vehículos | **Variables:** 33 columnas  
### Descripción
Este proyecto busca predecir si un vehículo adquirido en subasta
mayorista resultará ser una mala compra ("kick") — es decir, un auto
con defectos graves, título comprometido o problemas mecánicos que
impiden su reventa rentable.

## 1. Contexto de la Industria
La industria de **remarketing automotriz** gestiona la redistribución
de vehículos usados a través de subastas mayoristas (ADESA, Manheim).
Los dealers compran autos sin inspección completa, asumiendo el riesgo
de adquirir vehículos problemáticos conocidos como "kicks" o "lemons".

Con ~12% de malos registros sobre 73k transacciones, el costo
acumulado de estas malas compras representa pérdidas significativas
para los dealers en términos de costos de adquisición, traslado
y garantías no recuperables.

## 2. Objetivo de Negocio

Los dealers de autos usados que participan en subastas mayoristas
enfrentan pérdidas financieras al adquirir vehículos con defectos
graves, títulos comprometidos o problemas mecánicos no detectados
al momento de la compra.

El objetivo es desarrollar un modelo predictivo que, a partir de las
características disponibles en el momento de la subasta, permita al
dealer identificar con anticipación qué vehículos tienen alta
probabilidad de ser una mala compra — reduciendo pérdidas operativas
y mejorando la rentabilidad del portafolio de compras.

## 3. Variable Objetivo (Target)

| Campo      | Tipo    | Descripción                                      |
|------------|---------|--------------------------------------------------|
| `IsBadBuy` | Binaria | 1 = mala compra ("kick") / 0 = compra aceptable  |

**Distribución:**
- Clase 0 (buena compra): 64,007 registros → 87.7%
- Clase 1 (mala compra):   8,976 registros → 12.3%

**Dataset desbalanceado — requiere tratamiento especial en modelado.**

## 4. Preguntas de Negocio

1. ¿Este vehículo tiene alta probabilidad de ser una mala compra
   antes de adquirirlo en subasta?
2. ¿Qué características (marca, edad, odómetro, precio, etc) están más
   asociadas con ser un "kick"?
3. ¿Existen tipos de vehículos o casas de subasta donde el riesgo
   de mala compra es sistemáticamente mayor?
4. ¿El precio pagado por el dealer refleja adecuadamente el riesgo
   real del vehículo?

## 5. KPIs de éxito del Modelo

| KPI          | Métrica   | Umbral mínimo | Justificación                         |
|--------------|-----------|---------------|---------------------------------------|
| Discriminación | AUC-ROC | ≥ 0.75        | Evalúa separación entre clases        |
| Balance P/R  | F1-Score  | ≥ 0.50        | Penaliza falsos negativos y positivos |

**Restricciones del negocio:**
- El modelo usa solo variables disponibles *antes* de la compra
- Se prioriza **recall alto**: es más costoso comprar un kick
  que rechazar un buen auto
- Variables con >95% nulos (PRIMEUNIT, AUCGUART) serán excluidas
- No usar precios MMRCurrent (data leakage — son post-compra)

## 6. Glosario de Términos del Dominio

| Término        | Definición                                                  |
|----------------|-------------------------------------------------------------|
| Kick / Lemon   | Vehículo con problemas que impiden su reventa normal        |
| Bad Buy        | Sinónimo de kick; compra que genera pérdida al dealer       |
| Remarketing    | Proceso de redistribución de vehículos usados               |
| MMR            | Manheim Market Report — precio de referencia de mercado     |
| Dealer         | Revendedor de autos usados que compra en subasta            |
| ADESA/Manheim  | Principales casas de subasta automotriz en EE.UU.           |
| VehBCost       | Precio pagado por el dealer en la subasta                   |
| WarrantyCost   | Costo de garantía ofrecida al comprador final               |

## 8. Supuestos y Riesgos del Proyecto

### Supuestos
- Los datos reflejan fielmente las condiciones reales de subasta
- `IsBadBuy` fue etiquetado de forma consistente y confiable
- Las variables disponibles en subasta son suficientes para predecir
- El comportamiento histórico es representativo del futuro

### Riesgos

| Riesgo                          | Impacto | Mitigación                        |
|---------------------------------|---------|-----------------------------------|
| Desbalance de clases (88/12)    | Alto    | SMOTE / class_weight              |
| 95% nulos en PRIMEUNIT/AUCGUART | Medio   | Excluir o crear indicador de nulo |
| Data leakage con MMRCurrent     | Alto    | Usar solo MMRAcquisition          |
| Cambios en mercado automotriz   | Medio   | Reentrenamiento periódico         |
